```python
## import libraries for numpy, matplotlib and pandas

import matplotlib.pyplot as plt
import pandas as pd
```

```python
## read csv file and print in the workbook
earthquake_data = pd.read_csv('all_month.csv')
earthquake_data
```

```python
## pick the top 5 lines of the data to review
earthquake_data.head(5)
```

```python
## check the information about the data set
## if there are different types, we have to convert them into same data type 
earthquake_data.info()
```

```python
## check if there are any missing values
earthquake_data.isnull().sum()
```

```python
## 1st step: replace all the missing values w/ np.nan = placeholder for missing values
earthquake_data.replace(to_replace={col: {'':np.nan, None:np.nan} for col in earthquake_data.columns}, inplace=True)

## 2nd step: now replace all np.nan with median values for each column
earthquake_data['nst'] = earthquake_data['nst'].fillna(earthquake_data['nst'].median())
earthquake_data['gap'] = earthquake_data['gap'].fillna(earthquake_data['gap'].median())
earthquake_data['dmin'] = earthquake_data['dmin'].fillna(earthquake_data['dmin'].median())
earthquake_data['horizontalError'] = earthquake_data['horizontalError'].fillna(earthquake_data['horizontalError'].median())
earthquake_data['magError'] = earthquake_data['magError'].fillna(earthquake_data['magError'].median())
earthquake_data['magNst'] = earthquake_data['magNst'].fillna(earthquake_data['magNst'].median())

## 3rd step: after replacing the values, check to verify if there is still missing data
## data is now clean and can start working on it
earthquake_data.isnull().sum()
```

```python
earthquake_data.drop_duplicates(inplace=True)
earthquake_data
```

```python
round(earthquake_data, 2)
```

```python
## round the numbers to two decimal point as they are too long and difficult to read
round(earthquake_data.describe(), 2)
```

```python
earthquake_data.hist(bins=20, figsize=(20,10))
plt.show
```

```python
earthquake_data.hist('depth', bins=10, figsize=(5,5))
plt.show
```

```python
earthquake_data.hist('mag', bins=10, figsize=(5,5))
plt.show
```

```python
## make a boxplot of the data - what does it tell me?
## what are the outliers?
earthquake_data.boxplot(figsize=(15,10), flierprops = dict(marker='o', color='red'), medianprops = dict(linewidth=4))
earthquake_data.boxplot(medianprops= dict(color='black'))
plt.title('Earthquake Box Plot', loc='center')

```

```python
round(earthquake_data.groupby(['magSource']).max(), 2)
```

```python
## relation between depth vs magnitude
## use scatter plot
earthquake_data.plot(kind='scatter', x='mag', y='depth', c='depth', colormap='RdBu', alpha = 0.2)
plt.title('Earthquakes: Depth in Km and Magnitude', fontsize=16)
plt.gca().set_facecolor('none')
```

```python
## Create a pandas series for preparation of pie chart
## Counting earthquakes for each Magsource

quake_data = earthquake_data['magSource'].value_counts() 
quake_data
```

```python
pd.Series(quake_data[:5].plot(kind="bar"))
```

```python
## Create the pie chart for quake_data - top 5 area
colors = ['lightskyblue', 'red', 'blue', 'green', 'gold']

plt.pie(quake_data[:5], labels=quake_data[:5].index,colors=colors)

## label the pie chart with a title and where it will be located
plt.title('Pie Chart of Top 5 Earthquake Areas', loc="center")
plt.show()
```
